```python
from tifffile import imread
from glob import glob
import numpy as np
from matplotlib import pyplot as plt
```

```python
def get_index_arr(idx):
    idx_arr = np.array([])
    for i in idx:
        idx_arr = np.append(idx_arr, i.split('_')[-1].split('.')[0])
    return idx_arr
```

```python
def plot_image(x, y):

    plt.figure(figsize=(14, 7))

    plt.subplot(1,3,1).set_title('GT')
    plt.imshow(x)
    plt.subplot(1,3,2).set_title('Prediction')
    plt.imshow(y)
    plt.show()
```

## first bottom up layer

```python
labels_dir = glob('/group/jug/Sheida/maester_data/download/high_c1/label/1/*.tif')
labels = get_index_arr(labels_dir)
```

### centrioles

```python
centrioles_dir = glob('/group/jug/Sheida/maester_data/download/high_c1/centrioles/one_level/*.tif')
centrioles = get_index_arr(centrioles_dir)
```

```python
intersection_with_centrioles = np.intersect1d(labels, centrioles)
for i in intersection_with_centrioles[:1]:
    idx_label = np.where(labels == i)[0][0]
    idx_centrioles = np.where(centrioles == i)[0][0]

    y = imread(labels_dir[idx_label])
    x = imread(centrioles_dir[idx_centrioles])
    plot_image(x, y)
```
