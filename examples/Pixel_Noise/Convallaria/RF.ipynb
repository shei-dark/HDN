```python
import matplotlib.pyplot as plt
import numpy as np

# Define the size of the grid and the mask
grid_size = 64
mask_size = 4
kernel_size = 5

# Calculate the starting point of the mask
mask_start = (grid_size - mask_size) // 2

# Calculate the starting point of the kernel to overlap just one pixel with the top-left pixel of the red square
# The kernel's bottom-right corner should overlap with the top-left corner of the red square
kernel_start_i = mask_start + 1 - kernel_size 
kernel_start_j = mask_start + 1 - kernel_size + 2 * mask_size - 1

# Create a figure and axis with a specified aspect ratio to ensure no margins
fig, ax = plt.subplots(figsize=(8, 8))

# Draw the grid squares
for i in range(grid_size):
    for j in range(grid_size):
        # Highlight the central 4x4 region in red
        if mask_start <= i < mask_start + mask_size and mask_start <= j < mask_start + mask_size:
            color = 'red'
        else:
            color = 'white'
        
        # Draw the square
        square = plt.Rectangle((i, j), 1, 1, edgecolor='black', facecolor=color, linewidth=0.5)
        ax.add_patch(square)
        

# Draw the semi-transparent blue square representing the 5x5 kernel
kernel_rect = plt.Rectangle((kernel_start_i, kernel_start_j), kernel_size, kernel_size, edgecolor='blue', facecolor='blue', alpha=0.3, linewidth=2)
ax.add_patch(kernel_rect)

# Draw the semi-transparent blue square representing the 5x5 kernel
# Highlight the affected region in the output
affected_rect = plt.Rectangle((28, 28), 8, 8, edgecolor='green', facecolor='none', linewidth=2, linestyle='--')
ax.add_patch(affected_rect)

# Remove the axes for a cleaner look
ax.axis('off')

# Set the aspect of the plot to be equal to ensure squares
ax.set_aspect('equal')

# Set the limits to fit the grid exactly and remove any extra space
ax.set_xlim(0, grid_size)
ax.set_ylim(0, grid_size)

textstr = "Red: Masked 4x4 region\nBlue: 5x5 Kernel\nGreen: 8x8 RF"
props = dict(boxstyle='round', facecolor='yellow', alpha=0.8)
ax.text(32, 27, textstr, fontsize=10, verticalalignment='top', horizontalalignment='center', bbox=props)


# Remove the margin around the plot
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

# Hide the figure frame to remove the boundary
fig.patch.set_visible(False)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the size of the grid, mask, and receptive field
grid_size = 64
output_size = 32
mask_size = 12
kernel_size = 3
stride = 2
padding = 1

# Calculate the starting point of the mask in the input
mask_start = (grid_size - mask_size) // 2

# Calculate the corresponding affected region in the output
output_start = (mask_start + padding) // stride
output_end = (mask_start + mask_size - 1 + padding) // stride

# Create a figure and axis for the input image
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

# Draw the input image grid
for i in range(grid_size):
    for j in range(grid_size):
        color = 'white'
        if mask_start <= i < mask_start + mask_size and mask_start <= j < mask_start + mask_size:
            color = 'red'
        square = plt.Rectangle((i, j), 1, 1, edgecolor='black', facecolor=color, linewidth=0.5)
        ax[0].add_patch(square)

ax[0].set_xlim(0, grid_size)
ax[0].set_ylim(0, grid_size)
ax[0].set_aspect('equal')
ax[0].axis('off')
ax[0].set_title('Input Image')

# Draw the output image grid
for i in range(output_size):
    for j in range(output_size):
        color = 'white'
        if output_start <= i <= output_end and output_start <= j <= output_end:
            color = 'green'
        square = plt.Rectangle((i, j), 1, 1, edgecolor='black', facecolor=color, linewidth=0.5)
        ax[1].add_patch(square)

ax[1].set_xlim(0, output_size)
ax[1].set_ylim(0, output_size)
ax[1].set_aspect('equal')
ax[1].axis('off')
ax[1].set_title('Output Feature Map')

plt.show()
```
